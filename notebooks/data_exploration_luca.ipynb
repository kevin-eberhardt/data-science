```python
import sys
import os
from dotenv import load_dotenv

#1. load environment variables and data

# load environment variables
load_dotenv()

#add working directory to sys path to execute utils/dataset.py
working_dir = os.environ.get("WORKING_DIRECTORY")
sys.path.insert(0, working_dir)
```

```python
from utils.dataset import get_data 

df = get_data()
```

```python
#5. Check for missing values in the dataset using the isnull() method, and if any missing values exist, you can either impute or drop them.

df.isnull().sum()
```

```python
#6. Compute some summary statistics for the numerical features using the describe() method. This will give you an idea of the distribution of each feature, its mean, standard deviation, and quartile ranges.

df.describe()
```

```python
#7. Check for any correlations between the numerical features using a correlation matrix or a heatmap. 
#You can use the Pandas corr() method to compute the correlation matrix, and seaborn library to plot a heatmap.
#make it large enough to see the text
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, fmt='.1g')
```

```python
#8. Plot histograms or density plots for each numerical feature to see its distribution.

df.hist(figsize=(10,10))
```

```python
#9. Plot scatter plots to visualize the relationship between pairs of numerical features.

sns.pairplot(df)
```

```python
#10. Plot bar charts to visualize the distribution of the categorical feature

sns.countplot(x='quality', data=df)

```

```python
#11. Check for any outliers in the numerical features (fixed acidity, volatile acidity, citric acid, residual sugar, magnesium, flavanoids, minerals, calcium, chlorides, free sulfur dioxide, total sulfur dioxide, density, pH, sulphates, alcohol) using box plots.

#plot one boxplot for each numeric feature but show them in a grid
fig, axes = plt.subplots(5, 3, figsize=(15, 15))
for i, col in enumerate(df.select_dtypes('number')):
    sns.boxplot(ax=axes[i//3, i%3], x=col, data=df)

```

```python
#12. Finally, calculate the distribution of the labels (quality) using bar charts and explore the relationship between the features and the labels using scatter plots or box plots.

#calculate the distribution of the labels (quality)
sns.countplot(x='quality', data=df)

#explore the relationship between the features and the labels using scatter plots or box plots.
#plot one scatter plot for each numeric feature but show them in a grid
fig, axes = plt.subplots(5, 3, figsize=(15, 15))
for i, col in enumerate(df.select_dtypes('number')):
    sns.scatterplot(ax=axes[i//3, i%3], x=col, y='quality', data=df)

    
```

```python
#13 check the categorical feature (wine type) using a bar chart.
import seaborn as sns
sns.countplot(x='wine type', data=df)
```

```python
#14. explore the relationship between the categorical feature (wine type) and the labels (quality) using box plots. show only for wine type "Gamay"
sns.boxplot(x='wine type', y='quality', data=df[df['wine type']=='Gamay'])
```
