```python
#load csv in the same folder 
#filename: wine_test.csv

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

#load Meth csv
df_test = pd.read_csv('wine_test.csv', sep=';', decimal='.')

#load our csv
df_val = pd.read_csv('wine_validation_test.csv', sep=';', decimal='.')
```

### Meth Validation

```python
#compare statistics of the two datasets
df_test.describe()
#drop first column
df_test = df_test.drop(df_test.columns[0], axis=1)
df_test.describe()
```

### Our Validation

```python
df_val.describe()
```

```python
#calculate difference of statistics between the two datasets
df_test.describe() - df_val.describe()
```

```python
#calculate relative difference of statistics between the two datasets and put in heatmap

test = (df_test.describe() - df_val.describe())/df_test.describe()
#drop first row
test = test.drop(test.index[0])
#make it bigger
plt.rcParams['figure.figsize'] = [15, 8]
#plot heatmap
sns.heatmap(test, annot=True, cmap='coolwarm', linewidths=0.5)

```

```python
#plot magnesium boxplot for the two datasets
plt.rcParams['figure.figsize'] = [8, 3]
sns.boxplot(x='citric acid', data=df_test)
#sns.boxplot(x='magnesium', data=df_val)
```

```python
#plot magnesium boxplot for the two datasets
plt.rcParams['figure.figsize'] = [8, 3]
sns.boxplot(x='citric acid', data=df_val)
#sns.boxplot(x='magnesium', data=df_val)
```
