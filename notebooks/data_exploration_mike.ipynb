```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../wine_training.csv', delimiter = ";")
df.head()
```

```python
#Get an Overview of the Data: Take a look at the first few rows of the dataset using the .head() method to see what the data looks like.
#You can also use the .info() method to get more information about the dataset.
df.info()
```

```python
#Clean the Data: Clean the dataset by removing missing values, duplicates, or any irrelevant columns that will not be used in your analysis.
#Check for missing values in the dataset.
df.isnull().sum()
df

#If there are any missing values, remove them
df = df.dropna()
df
```

```python
#Visualize the Data: Use visualizations to explore the dataset and gain insights. Matplotlib and Seaborn are great libraries for creating visualizations in Python.
#Create a histogram of the target variable (quality) to see how many wines are in each category.
sns.countplot(df['quality'])
```

```python
#Check for any correlations between the numerical features using a correlation matrix or a heatmap. 
#You can use the Pandas corr() method to compute the correlation matrix, and seaborn library to plot a heatmap.
#make it large enough to see the text
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, fmt='.0%')
```

```python
#Analyze the Data: Analyze the dataset by calculating descriptive statistics, creating pivot tables, or using machine learning algorithms.
#Calculate the average amount of all ingredients for each quality level of wine.
df.groupby('quality').mean()
df

# Create a pivot table
pivot = df.pivot_table(index=['quality'], values = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol'])
pivot
```
