```python
#EDA - Exploratory Data Analysis - Data Exploration - Data Visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def eda(dataframe):
    # Shape of the data
    print("Number of rows: ", dataframe.shape[0])
    print("Number of columns: ", dataframe.shape[1])
    
    # Summary statistics
    print("\nSummary Statistics:")
    print(dataframe.describe())
    
    # Missing values
    print("\nMissing Values:")
    print(dataframe.isnull().sum())
    
    # Categorical features
    print("\nCategorical Features:")
    print(dataframe.select_dtypes(include=['object']).nunique())
    
    # Histogram for all numeric columns
    # make it large enough to see the plot clearly
    dataframe.hist(figsize=(20, 20), bins=50, xlabelsize=8, ylabelsize=8)
    
    # Correlation matrix
    # make it large enough to see the plot clearly
    plt.figure(figsize=(20, 20))
    sns.heatmap(dataframe.corr(), annot=True, fmt=".2f")
    plt.show()

eda(df)
```
